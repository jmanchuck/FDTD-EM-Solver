```python
%matplotlib inline

from solver import Solver
from analyser import FileLoader as FL
import matplotlib.pyplot as plt
import numpy as np
```

```python
# initiate variables
sigma_w = 5 * 10 ** 8  # frequency bandwidth
omega_0 = 2 * 10 ** 9  # central frequency
s = 20  # mesh points per wavelength
stability = 1 / 10  # time mesh stability factor
C = 2.99 * (10 ** 8)

wavelength = 2 * np.pi * C / omega_0  # c = f * lambda, f = omega / 2pi

simulation_size = 4 
simulation_time = 10 * simulation_size / C

print("Simulation time (seconds):", simulation_time)
```

```python
# initiate solver with user input variables
solver = Solver(points_per_wavelength=s, stability=stability, eps_r_max=3.5, mu_r_max=1, simulation_size=simulation_size, simulation_time= simulation_time)

solver.set_reflect_boundaries(False, False, False, False)

pulse = solver.add_sinusoidal_pulse(omega_0, (0.7, 0))
# pulse = solver.add_sinusoidal_pulse(omega_0, (0.7, 0))
pulse.plot_time(scaled_plot=False)
```

```python
from solver import MaterialArray
material = MaterialArray(1000, 4, 4)

class solver:
    length_x = 4 
    length_y = 4

from_edge = 0.3
thickness = 0.4
rect_length = 0
eps = 5

h_rect_upleft = (from_edge, 0)
h_rect_botright = (from_edge + thickness, rect_length)

v_rect_upleft = (solver.length_y - rect_length, solver.length_x - from_edge - thickness)
v_rect_botright = (solver.length_y, solver.length_x - from_edge)

circle_center = (v_rect_upleft[0], h_rect_botright[1])
outer_radius = solver.length_x - from_edge - rect_length


material.set_material_rect(h_rect_upleft, h_rect_botright, eps)
material.set_material_rect(v_rect_upleft, v_rect_botright, eps)
material.set_quarter_circle(circle_center, outer_radius, eps, thickness=thickness)


material.plot()
```
